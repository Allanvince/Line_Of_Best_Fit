```python
import numpy as np
```

```python
from scipy.optimize import curve_fit
```

```python
Year,Mean = np.loadtxt("annual_Mean_Temperatures.txt", unpack = True, usecols=(1,2),skiprows=1)
```

```python
Mean.min()
```

```python
Mean.max()
```

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(Year,Mean,'.')
plt.xlabel('Year')
plt.ylabel('Annual Average Temperature Temperature')
```

```python
e = np.repeat(0.2,274)
e
```

```python
def line(x,a,b):
    return a*x + b
```

```python
popt, pcov = curve_fit(line,Year,Mean,sigma=e)
```

```python
popt
```

```python
pcov
```

```python
plt.errorbar(Year,Mean,yerr=e,fmt='.')
```

```python
plt.errorbar(Year,Mean,yerr=e, fmt='.',color = 'green')
xfine = np.linspace(1880.0,2016.0,274)
plt.plot(xfine,line(xfine,*popt),'r-')
```
